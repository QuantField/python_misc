```python
# Simplest Neural Net for regression: one layer, one node, one ouput.

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(3405)
# generating  the data 
x  = np.linspace(-5,5,100)
yt = 3*x+1 
y  = yt + 5*np.random.randn(len(x))

# initial random w,b
w = 2*np.random.rand()-1
b = 2*np.random.rand()-1

def pred(x):
    return w*x+b

eta = 0.001 # learning rate
s = 1e20
for k in range(300):
    err = 0
    for i in range(len(y)):
        dy = (pred(x[i])-y[i])
        w, b = w-eta*dy*x[i], b-eta*dy
        err += dy*dy   
    s0, s = s, err    
    if s>s0: break
    print("iter :", k, "SSE= ",err,"w= ",w," b= ",b)

plt.plot(x,y,'o', label='noisy')
plt.plot(x,pred(x),'-k', label='predicted')
plt.plot(x,yt,'-r', label='actual')
plt.legend()
plt.grid()
plt.show()
```
