```python
"""
two layers with tanh as activation functions
Cost function is the SSE

"""


class layer:
    def __init__(self, n_nodes, n_inputs):
        self.n_nodes  = n_nodes
        self.n_inputs = n_inputs
        self.w = 2*np.random.random([n_nodes, n_inputs])-1
    def feed(self,x):
        return  np.tanh(np.dot(self.w,x.reshape(-1,1)))
    def feedData(self,X):
        d = np.dot(X,self.w.T)
        return np.tanh(d)
    
eta  = 0.01    
n_nodes = 20             
in_dim = X.shape[1]
np.random.seed(234) 
L1    = layer(n_nodes ,in_dim)
L2    = layer(1, n_nodes)
iters = 1000
SSE = np.Infinity
for i in range(iters):    
    err  = [] 
    for obs in range(X.shape[0]):
        o0 = X[obs,:]
        o1 = L1.feed(o0)
        o2 = float(L2.feed(o1))
        
        y = float(Y[obs])
        dE_do2 = -(y-o2)
        
        err.append(y-o2)
        delta2 = dE_do2*(1-o2*o2)
        L2.w  += -eta*delta2*o1.reshape(1,-1)
            
        delta1 = delta2*(1-o1*o1)    
        L1.w += -eta*delta1*o0
    SSE0 = SSE    
    SSE  = np.dot(err,err)    
    if SSE>SSE0 and i>50: 
        break    
    print("epoch = ",i," SSE =",SSE)
    

O1    = L1.feedData(X)
O2    = L2.feedData(O1)
pred  = O2.copy()
pred[pred<0]=-1
pred[pred>0]=1
    

acc = sum([pred[i]!=Y[i] for i in range(len(Y))])/len(Y)
print("\ntrain acc =",float(acc))

from sklearn.metrics import precision_recall_fscore_support, confusion_matrix 
conf = confusion_matrix(Y, pred) 
confMat = conf[0], conf[1]    
[TN, FN], [FP, TP] = confMat
print('TN,  FN,  FP,  TP')
print(TN,  FN,  FP,  TP)

import sklearn.metrics as sk
fpr, tpr, thresholds = sk.roc_curve(Y, pred)
auroc = sk.auc(fpr, tpr)
```
