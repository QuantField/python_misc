```python
import tensorflow as tf
import numpy as np
import pandas as pd

np.random.seed(11)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.utils import shuffle
```

```python
# Data 
#-----------
path  = "C:\\myData\\banana\\"
df = pd.read_csv(path+"banana_train_1.csv", header = None)
n,m = df.shape
X = df[df.columns[0:m-1]].values
y = df[df.columns[m-1]]
encoder = LabelEncoder().fit(y)
Y = encoder.transform(y)
Y2 = 1- Y
# the label Y is hotcoded 2 values instead of 1
# istead of 1 it is [1,0] and instead of 0 it is [0,1]
Y = np.hstack((Y.reshape(-1,1),Y2.reshape(-1,1)))
X, Y = shuffle(X,Y, random_state=23)
train_x, test_x, train_y, test_y = train_test_split(X,Y, test_size=0.30, random_state = 3495)
#print(test_x.shape, train_x.shape)
```

```python
n_class = 2
n_dim   = X.shape[1]
mod_path= "C:\\Users\\ks_work\\Desktop\\Code\\tensFlow_py\\MP\\"

eta = 0.1
n_hidden_1 = 60
n_hidden_2 = 30
n_hidden_3 = 10

x  = tf.placeholder(tf.float32, [None, n_dim]) # None means anything 
y_ = tf.placeholder(tf.float32, [None, n_class])

#w  = tf.Variable(tf.zeros([n_dim,n_class]))
#b  = tf.Variable(tf.zeros[n_class])


def mlp(x, weights, biases):
    
    layer1 = tf.add(tf.matmul(x,weights['h1']), biases['b1'])
    layer1 = tf.nn.sigmoid(layer1)
    
    layer2 = tf.add(tf.matmul(layer1,weights['h2']), biases['b2'])
    layer2 = tf.nn.sigmoid(layer2)
    
    layer3 = tf.add(tf.matmul(layer2,weights['h3']), biases['b3'])
    layer3 = tf.nn.relu(layer3)
    
    out_layer = tf.matmul(layer3,weights['out'])+ biases['out']
    
    return out_layer
    
weights ={
    'h1' : tf.Variable(tf.truncated_normal([n_dim, n_hidden_1])),
    'h2' : tf.Variable(tf.truncated_normal([n_hidden_1, n_hidden_2])),
    'h3' : tf.Variable(tf.truncated_normal([n_hidden_2, n_hidden_3])),
    'out': tf.Variable(tf.truncated_normal([n_hidden_3, n_class]))  
}
    
biases ={
    'b1' : tf.Variable(tf.truncated_normal([n_hidden_1])),
    'b2' : tf.Variable(tf.truncated_normal([n_hidden_2])),
    'b3' : tf.Variable(tf.truncated_normal([n_hidden_3])), 
    'out': tf.Variable(tf.truncated_normal([n_class]))
}    
    
init = tf.global_variables_initializer()   

saver = tf.train.Saver()

y = mlp(x, weights, biases)

cost_func   = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y, labels=y_))
train_steps = tf.train.GradientDescentOptimizer(eta).minimize(cost_func)

sess = tf.Session()
sess.run(init)

mse = []
acc = []
err = []

train_epochs = 1000

for epoch in range(train_epochs):
    sess.run(train_steps, feed_dict={x: train_x, y_:train_y})
    cost = sess.run(cost_func, feed_dict={x: train_x, y_:train_y} )
    err.append(cost)
    # y and y_ are 2 column vecs, correct prediction means
    # the presence of 1 and 0 in the correct place
    # hence argmax
    corr_pred = tf.equal(tf.argmax(y,1),tf.argmax(y_,1))
    accu      = tf.reduce_mean(tf.cast(corr_pred,tf.float32))
    pred_y    = sess.run(y, feed_dict={x:test_x})
    mse_      = sess.run( tf.reduce_mean(tf.square(pred_y-test_y)))
    mse.append(mse_)
    acc_ = sess.run(accu, feed_dict={x: train_x, y_:train_y} )
    acc.append(acc_)
    print(epoch, cost, acc_)
    
```
