# Data manipulation

```python
import pandas as pd
import numpy  as np
```

```python
np.random.seed(25)
N = 10
df = pd.DataFrame({'V1': np.random.randn(N), 'V2':np.random.randn(N)})
df
```

```python
# changing values using indices
df2 = df.copy() # deep copy
df2.iloc[2:5,0] = 1 # first column 0 , rows from 2 t0 4 = 1
print(df)
print(df2)
```

```python
# using labeld or variable names
df2 = df.copy()
df2.loc[3:5, 'V2'] = np.nan
df2.loc[(df2['V1']<0),'V1'] = 0
df2
```

```python
M = N*5
tab = np.random.randn(M).reshape(N,5)
varNames = ['Var1','Var2','Var3','Var4','Var5'] 
df = pd.DataFrame(tab, columns=varNames)
df  
```

```python
df.loc[3:4,'Var1']= np.nan
df.loc[1:3,'Var4']= np.nan
df
```

```python
mean = df.mean()  
mean
mnDict = dict(mean)
mnDict
```

```python
# filling missing values with the mean 
df.fillna(mean, inplace=True) # using the dictionary mnDict will work as well
df
```
